```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import statsmodels.formula.api as smf
from scipy import stats
```

```python
league = {'brasileirao_a': ['Brasileirao'],
          'premier_league': ['Premier_League'],}


liga = league['brasileirao_a'][0]
```

```python
historico = pd.read_excel(f'../datasets/{liga}/match_history/all_games/historico-2015a2023.xlsx')
```

```python
historico.head()
```

# Modelo

```python
modelo = smf.glm(formula= 'gf ~ team + venue + opponent', data= historico, family= sm.families.Poisson()).fit()
```

```python
print(modelo.summary())
```
